```python
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dp= pd.read_csv('colors.csv')
```

```python
image_path = 'colorpic.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for displaying with matplotlib

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
lower_red = np.array([0, 0, 100])  # Lower bound for red color in BGR format
upper_red = np.array([100, 100, 255])  # Upper bound for red color in BGR format
```

```python
mask = cv2.inRange(image, lower_red, upper_red)
result = cv2.bitwise_and(image_rgb, image_rgb, mask=mask)

plt.imshow(result)
plt.axis('off')
plt.show()
```

```python
detected_color = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
plt.imshow(detected_color)
plt.axis('off')
plt.show()
```
